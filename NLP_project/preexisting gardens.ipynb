```python
import os 

import numpy as np
import pandas as pd
import folium as folium
import requests
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import json
import folium
from folium import plugins
```

```python
preexisting = pd.read_csv('gardens.locations.csv',encoding='ISO-8859-1')
preexisting
```

```python
preexisting = preexisting.dropna()

preexisting.head(20)
```

```python
import folium
from folium.plugins import MarkerCluster

# Starting coordinates
la = [34.0522, -118.2437]

# Create a Folium map object with Los Angeles as the default location
gardenmap = folium.Map(location=la, zoom_start=10)  # Adjust the zoom level as needed

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(gardenmap)

# Filter out rows with null values for latitude and longitude
df_filtered = preexisting.dropna(subset=['Lat', 'Long'])

# Iterate through the filtered DataFrame and add markers for each latitude and longitude pair
for index, row in df_filtered.iterrows():
    folium.Marker([row['Lat'], row['Long']]).add_to(marker_cluster)

gardenmap
```

```python
gardenmap.save("gardenmap.html")
```

```python
#known empty lots of land in los angeles (pulled from zillow)
df = pd.read_csv('zillow2.csv')
df.head(50)
```

```python
import folium
from folium.plugins import MarkerCluster

# Starting coordinates
la = [34.0522, -118.2437]

# Create a Folium map object with Los Angeles as the default location
map = folium.Map(location=la, zoom_start=8)  # Adjust the zoom level as needed

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(map)

# Filter out rows with null values for latitude and longitude
df_filtered = df.dropna(subset=['latitude', 'longitude'])

# Iterate through the filtered DataFrame and add markers for each latitude and longitude pair
for index, row in df_filtered.iterrows():
    folium.Marker([row['latitude'], row['longitude']]).add_to(marker_cluster)

map
```

```python
from geopy.geocoders import Nominatim

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="Geopy Library")

# Address to geocode
address = "4712 S. Vermont Avenue\nLos Angeles, CA 90037	"

# Geocode the address
location = geolocator.geocode(address)

# Check if location was found
if location:
    latitude = location.latitude
    longitude = location.longitude
    print(f"Latitude: {latitude}, Longitude: {longitude}")
else:
    print("Location not found.")
```
